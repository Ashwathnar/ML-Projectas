```python
#split the "int" feature and "object" feature

objch=[]
numin=[]
for gh in df.columns:
    if df[gh].dtype=="object":
        objch.append(gh)
    else:
        numin.append(gh)
```

```python
# split nominal charecters and ordinal charecters feature

targete=['brand','processor_brand','processor_name','processor_gnrtn','os','weight']
labele=['ram_gb','ram_type','ssd','hdd','os_bit','graphic_card_gb','warranty','Touchscreen','msoffice','rating']
```

```python
#import encoders model

from sklearn.preprocessing import LabelEncoder
from category_encoders import TargetEncoder
```

```python
# ordinal charecter feature values change to numerical values using "label encoder"
#(# nomial charecter feature values change to numerical value  using  "Target encoder"
# because this feature have lot of unique charecter)

for ta in targete:
    te=TargetEncoder()
    df[ta]=te.fit_transform(df[ta],df['Price'])
for la in labele:
    lab=LabelEncoder()
    df[la]=lab.fit_transform(df[ta])
    
```

```python
df
```

```python
# FIND THE CORRELATION BETWEEN TARGET TO INDEPENDENT FEATUREs

corre=abs(df.corr()['Price'])
corre
```

```python
#filter the porr correlation feature to target feature

badcorr=corre[corre<0.25]
badcorr
```

```python
#drop the bad correlation features

df.drop(badcorr.index,axis=1,inplace=True)
```

```python
df.shape
```

```python
X=df.drop('Price',axis=1)
y=df['Price']
```

```python
#import regression model algorithums

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
# import evalation metrics
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
from sklearn.linear_model import Ridge
# import ensemble  stacking metod to use increase accureasy 
from sklearn.ensemble import StackingRegressor


X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
#train the model and predict the values then find the accuracy of mae,mse,r2_score

models=[LinearRegression(),DecisionTreeRegressor(),RandomForestRegressor(),SVR()]
for mo in models:
    mod=mo
    mod.fit(X_train,y_train)
    pre=mod.predict(X_test)
    mae=mean_absolute_error(pre,y_test)
    mse= mean_squared_error(pre,y_test)
    r2=r2_score(pre,y_test)
    print()
    print(mo)
    print("mean_absolute_error",mae)
    print()
    print("mean_squared_error",mse)
    print()
    print("r2_score",r2)
    print()
    
```

```python
# RandomForestRegressor() and DecisionTreeRegressor() is a good traind model
```

```python
# increase accurasy i using stackingregressor


# Define base models
base_models = [
    ('lr', LinearRegression()),
    ('rf', RandomForestRegressor(random_state=42)),
    ('svr', SVR()),
    ('dc',DecisionTreeRegressor())
]

# Initialize the StackingRegressor with the base models and a meta-model
stacked_model = StackingRegressor(estimators=base_models, final_estimator=Ridge())

# Fit the stacked model on the training data
stacked_model.fit(X_train, y_train)

# Evaluate the stacked model on the testing data
score = stacked_model.score(X_test, y_test)
print(f"Stacked model R^2 score: {score:.3f}")
print(f"staked model mean_squared_error: {score:}")
print(f"staked mean_absolute_error:{score:}")
```
