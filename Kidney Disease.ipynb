```python
#import nessary module
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#using pandas and read dataset
df=pd.read_csv("kidney_disease.csv")
df
```

```python
df.drop("id",axis=1,inplace=True)
```

```python
#feature is short form,i change full form of the features

df=df.rename(columns={'id': 'Identification',
    'age': 'Age',
    'bp': 'Blood Pressure',
    'sg': 'Specific Gravity',
    'al': 'Albumin',
    'su': 'Sugar',
    'rbc': 'Red Blood Cells',
    'pc': 'Pus Cell',
    'pcc': 'Pus Cell Clumps',
    'ba': 'Bacteria',
    'bgr': 'Blood GlucoseRandom',
    'bu': 'Blood Urea',
    'sc': 'Serum Creatinine',
    'sod': 'Sodium',
    'pot': 'Potassium',
    'hemo': 'Hemoglobin',
    'pcv': 'Packed Cell Volume',
    'wc': 'White Blood Cell Count',
    'rc': 'Red Blood Cell Count',
    'htn': 'Hypertension',
    'dm': 'Diabetes Mellitus',
    'cad': 'Coronary Artery Disease',
    'appet': 'Appetite',
    'pe': 'Pedal Edema',
    'ane': 'Anemia',
    'classification': 'Classification'})
```

```python
# Feature store numbers but feature datatype is object,this feature has wrong datatype ,i change the datatype
#using pandas and to_numeric() function

numobj=['Packed Cell Volume','White Blood Cell Count','Red Blood Cell Count']
for nu in numobj:
    df[nu]=pd.to_numeric(df[nu],errors="coerce")
```

```python
# this dataset features has null values in "object"datatype and "int "datatype
#so i seprate the object dataype feature and intdatatype features

numf=[]
objectf=[]
for h in df.columns:
    if df[h].dtype=='object':
       objectf.append(h)
    else:
        numf.append(h)
    
```

```python
numf
```

```python
objectf
```

```python
#create a two function ,the function is mean and mode 
# mean means the feature mean value take and add null values place using 'fillna()'
# mode means which is more charecter present in the feature and that first index[0] charecter take and fill the 
# nullvalues in the object dtype features
def mean (df,columns):
    meana =df[columns].mean()
    df[columns].fillna(meana,inplace=True)
    
def mode (df,columns):
    modea =df[columns].mode()[0]
    df[columns].fillna(modea,inplace=True)    
    
    


for g in numf:
   mean(df,g)
for f in objectf:
   mode(df,f)
```

```python
# check the how many chategries in the objecet datatypes

for u in objectf:
    g=df[u].unique()
    print(u +": ")
    print(g)
    print( )
    
```

```python
# i changed the wrong chategries
df['Diabetes Mellitus'].replace({' yes':'yes', '\tno':'no', '\tyes':'yes'},inplace=True)
df['Coronary Artery Disease'].replace({'\tno':'no', },inplace=True)
df['Classification'].replace({'ckd\t':'ckd'},inplace=True)
```

```python
print(df['Diabetes Mellitus'].unique())
print(df['Coronary Artery Disease'].unique())
print(df['Classification'].unique())
```

```python
#now i change object chategries to number chategries  becase ml easily understan and accurately find the patterns
# i using label encoder because all features has two type of charecterstics

from sklearn.preprocessing import LabelEncoder

for d in objectf:
    la=LabelEncoder()
    df[d]=la.fit_transform(df[d])
```

```python
# annalysing relationship with one feature to another feature

import seaborn as sns
corr_matrix=df.corr()
plt.figure(figsize=(10, 15))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix Heatmap Data")
plt.show()
```

```python
# i find the realtionship with target to independent variable

rela = abs(df.corr()['Classification'].sort_values(ascending=False))
rela
```

```python
# i fillter the poor relationship feature with target

filtered_corr = rela[rela< 0.34]
print(filtered_corr)
```

```python
# filltered poor relationship feature drop

df.drop(filtered_corr.index, axis=1, inplace=True)
```

```python
#seprate test and train
X=df.iloc[:,:13]
y=df.iloc[:,13]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
# import classification algorithum
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
#find the accuracy and confussion matrix of the algorithums

model=[DecisionTreeClassifier(),RandomForestClassifier(),KNeighborsClassifier(),SVC()]
for fun in model:
    cla=fun
    cla.fit(X_train,y_train)
    pre=cla.predict(X_test)
    accs= accuracy_score(pre,y_test)
    con=confusion_matrix(pre,y_test)
    
    print(fun,":")
    print("accuracy_score is   : ",accs)
    print("confusion matrix is :")
    print(con)
    print()

    
    
    
    
```
