### 1.Encoding categorical varibles


```python
#find frequency of each Cuisines values
fre=df['Cuisines'].value_counts()
#map frequency in the dataset
df['Cuisinesen']=df['Cuisines'].map(fre)
```

```python
la=['Has Table booking', 'Has Online delivery',
       'Is delivering now', 'Switch to order menu', 'Rating color',
       'Rating text']
ta=['Restaurant Name', 'City', 'Address', 'Locality', 'Locality Verbose',
       'Cuisines', 'Currency']
```

```python
from sklearn.preprocessing import LabelEncoder
import category_encoders as ce
L=LabelEncoder()
T=ce.TargetEncoder()
for i in la:
    df[i]=L.fit_transform(df[i])
for i in ta:
    df[i]=T.fit_transform(df[i],df['Cuisinesen'])
```

```python
dro=[ 'Restaurant ID', 'Is delivering now',
       'Average Cost for two', 'Has Online delivery', 'Rating text',
       'Has Table booking', 'Votes', 'Country Code', 'Price range',
       'Aggregate rating', 'Switch to order menu']
df.drop(dro,axis=1,inplace=True)
```

### 2.split data into training and testing


```python
#split feature and label
X=df.drop('Cuisinesen',axis=1)
y=df['Cuisinesen']
```

```python
#split train and test
from sklearn.model_selection import train_test_split
X_train,X_test,y_tarin,y_test=train_test_split(X,y,test_size=0.25,random_state=45)
```

```python
#scaling feature
from sklearn.preprocessing import StandardScaler
s=StandardScaler()
X_train=s.fit_transform(X_train)
X_test=s.fit_transform(X_test)
```

### 3.Build the model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_score,recall_score,accuracy_score
m=[LogisticRegression(max_iter=1000),RandomForestClassifier()]
for i in m:
    i.fit(X_train,y_train)
    pred=i.predict(X_test)
    pre=precision_score(pred,y_test,average='macro',zero_division=0)
    rec=recall_score(pred,y_test,average='macro',zero_division=0)
    acc=accuracy_score(pred,y_test)
    print(i)
    print('precision_score : ',pre)
    print('recall_score : ',rec)
    print('accuracy_score : ',acc)
    print()
```

i got randomforest 59% and logistic 45% accuracy ,althrough random forest better than logistic but it is not good ,i will use cross validation for increasing model accuracy

### 4.Build cross validation

```python
from sklearn.model_selection import cross_validate
from sklearn.pipeline import make_pipeline
from sklearn.metrics import make_scorer

rf=RandomForestClassifier()
pipe=make_pipeline(StandardScaler(),rf)

cv=cross_validate(pipe,X_train,y_train,cv=5,
                 scoring={'acc':'accuracy',
                         'precision':make_scorer(precision_score,average='macro',zero_division=0),
                         'recall':make_scorer(recall_score,average='macro',zero_division=0)},
                 return_train_score=False)
print('cross_validate')
print(f"Accuracy: {cv['test_acc'].mean():.4f}±{cv['test_acc'].std():4f}")
print(f"Precision: {cv['test_precision'].mean():.4f}±{cv['test_precision'].std():4f}")
print(f"Recall: {cv['test_recall'].mean():.4f}±{cv['test_recall'].std():4f}")
```
